```python
import json
import pandas as pd
```

### House Committee Data

```python
f = open('house_committees.json')
```

```python
data = json.load(f)
```

```python
data_simple = []
```

```python
for i in data:
    print(i)
    data_simple.append(i)
```

```python
data_simple
```

```python
data['results'][0]['committees']
```

```python
df = pd.DataFrame(data['results'][0]['committees'])
```

```python
display(df)
```

```python
df_2 = pd.json_normalize(data['results'][0]['committees'], max_level = 4)
display(df_2.head(5))
print(df_2.shape)

df_2.to_csv("house_committees.csv")

# print(df_2["subcommittees"].head(5))
df_sub = pd.json_normalize(df_2["subcommittees"][0])
display(df_sub)
```

```python
sub_df = pd.DataFrame()
for i in df_2["subcommittees"]:
    sub_df = pd.concat([sub_df,pd.json_normalize(i)], axis = 0)
    
sub_df.shape
sub_df.head(5)
```

```python
def return_four(row):
    return row['id'][:4]

sub_df['col_join'] = sub_df.apply(lambda row: return_four(row), axis = 1)

sub_df.head(5)
```

```python
house_committees_v2 = df_2.merge(sub_df, how = 'outer', left_on = 'id', right_on = 'col_join')
house_committees_v2.drop("subcommittees", axis = 1, inplace = True)
house_committees_v2.to_csv("house_committees_v2.csv")
```

### Senate Committee Data

```python
f = open('senate_committees.json')
data_3 = json.load(f)
df = pd.DataFrame(data_3['results'][0]['committees'])
df.head(5)
```

```python
sub_df = pd.DataFrame()
for i in df["subcommittees"]:
    sub_df = pd.concat([sub_df,pd.json_normalize(i)], axis = 0)

sub_df['col_join'] = sub_df.apply(lambda row: return_four(row), axis = 1)
    
senate_committees_v2 = df.merge(sub_df, how = 'outer', left_on = 'id', right_on = 'col_join')
senate_committees_v2.drop("subcommittees", axis = 1, inplace = True)
senate_committees_v2.to_csv("senate_committees_v2.csv")
```
