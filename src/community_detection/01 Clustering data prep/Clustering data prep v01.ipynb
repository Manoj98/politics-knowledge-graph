### Extracting bill numbers from voting information

```python
HOUSE_BILL_KEYWORDS = [" H R ", " H RES ", " H.R. ", " H.Res. ", " H.J.Res. ", " H.J. Res", " H J RES "]
```

```python
SENATE_BILL_KEYWORDS = [" S ", " S. ", " S.Res. ", " S.J.Res. ", " S.J. Res ", " S J RES "]
#"S.", "S.J."
```

```python
def extract_bill_id_from_vote(vote_question):
    #print(vote_question)
    bill_keywords = HOUSE_BILL_KEYWORDS + SENATE_BILL_KEYWORDS
    matched_keyword, matched_keyword_index = None, len(vote_question)
    for keyword in bill_keywords:
        if keyword in vote_question:
            keyword_index = vote_question.index(keyword)
            if keyword_index < matched_keyword_index:
                matched_keyword_index = keyword_index
                matched_keyword = keyword
    if matched_keyword is None:
        return None, None
    i = matched_keyword_index + len(matched_keyword)
    j = i + 1
    while j < len(vote_question) and vote_question[i:j].isnumeric():
        j += 1
    #print(matched_keyword, vote_question[i:j])
    try:
        return matched_keyword, int(vote_question[i:j])
    except ValueError:
        return None, None
```

```python
def row_addition_1(row):
#     kw, n = extract_bill_id_from_vote(row["question"])
    kw, n = extract_bill_id_from_vote(row)
    return kw

def row_addition_2(row):
    kw, n = extract_bill_id_from_vote(row)
    return n
```

```python
df_senate_votes_copy = df_senate_votes.copy()
df_house_votes_copy = df_house_votes.copy()

df_house_votes_copy['kw'] = np.nan
df_house_votes_copy['n'] = np.nan

df_house_votes_copy['kw'] = df_house_votes_copy["question"].apply(row_addition_1)
df_house_votes_copy['n'] = df_house_votes_copy["question"].apply(row_addition_2)
```

```python
df_house_votes_copy.to_clipboard()
```

```python
def cleanup(row):
    try:
        kw_clean = row['kw'].strip().lower().replace(" ", "").replace(".", "")
        n = int(row['n'])
#         print(kw_clean+str(n))
    except:
        kw_clean = str(np.nan)
        n = str(np.nan)
    
    return kw_clean+str(n)
```

```python
df_house_votes_copy['combined_bill'] = df_house_votes_copy.apply(lambda row: cleanup(row), axis = 1)
```

```python
df_house_votes_copy['combined_bill_join'] = df_house_votes_copy['combined_bill'] + "-116"
```
