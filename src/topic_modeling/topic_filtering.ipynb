```python
import pandas as pd
```

```python
all_topics = pd.read_csv("../../data/topics/subject_topic_full_edges.tsv", sep="\t")
all_topics.head()
```

```python
all_topics[all_topics["top_name"] == "Economics and public finance"]
```

```python
keep_topics = ["Government operations and politics", "Finance and financial sector", "Economics and public finance",
                "Armed forces and national security", "Health"]

keep_topics_df = all_topics[all_topics["top_name"].isin(keep_topics)].reset_index()
del keep_topics_df["index"]
```

```python
gov_keep_subjects = ["Government operations and politics", "Government information and archives",
                 "political campaign regulation", "Elections", "voting", "State and local government operations",
                 "benefits", "personnel management", "Government employee pay", "public corruption"]

gov_keep_df = keep_topics_df[keep_topics_df["sub_name"].isin(gov_keep_subjects)]
gov_keep_df = gov_keep_df[gov_keep_df["top_name"] == "Government operations and politics"].reset_index()
del gov_keep_df["index"]
gov_keep_df = gov_keep_df.drop_duplicates(subset=["sub_name"])
```

```python
finance_keep_subjects = ["Finance and financial sector", "Government studies and investigations",
                        "Congressional oversight", "Government information and archives", "Securities",
                        "Administrative law and regulatory procedures", "Financial services and investments",
                        "Banking and financial institutions regulation", "Business records", "Fraud offenses and financial crimes"]
finance_keep_df = keep_topics_df[keep_topics_df["sub_name"].isin(finance_keep_subjects)]
finance_keep_df = finance_keep_df[finance_keep_df["top_name"] == "Finance and financial sector"].reset_index()
finance_keep_df = finance_keep_df.drop_duplicates(subset=["sub_name"])
del finance_keep_df["index"]
```

```python
eco_keep_subjects = ["Economics and public finance", "Executive agency funding and structure", "Appropriations",
                    "Department of Health and Human Services", "Congressional oversight", "Indian social and development programs",
                    "Government trust funds", "Department of Housing and Urban Development", "Roads and highways", "District of Columbia"]
eco_keep_df = keep_topics_df[keep_topics_df["sub_name"].isin(eco_keep_subjects)]
eco_keep_df = eco_keep_df[eco_keep_df["top_name"] == "Economics and public finance"].reset_index()
eco_keep_df = eco_keep_df.drop_duplicates(subset=["sub_name"])
del eco_keep_df["index"]
```

```python
natsec_keep_subjects = ["Armed forces and national security", "Congressional oversight",
                        "Veterans medical care", "Asia", "Conflicts and wars", 
                        "Government studies and investigations", "Government information and archives",
                        "Military operations and strategy", "Executive agency funding and structure"]
natsec_keep_df = keep_topics_df[keep_topics_df["sub_name"].isin(natsec_keep_subjects)]
natsec_keep_df = natsec_keep_df[natsec_keep_df["top_name"] == "Armed forces and national security"].reset_index()
natsec_keep_df = natsec_keep_df.drop_duplicates(subset=["sub_name"])
del natsec_keep_df["index"]
```

```python
health_keep_subjects = ["Health", "Health programs administration and funding", "Congressional oversight",
                        "Health promotion and preventive care", "Prescription drugs", "Health care costs and insurance",
                        "Department of Health and Human Services", "Administrative law and regulatory procedures", 
                        "Government information and archives", "Government studies and investigations"]
health_keep_df = keep_topics_df[keep_topics_df["sub_name"].isin(health_keep_subjects)]
health_keep_df = health_keep_df.drop_duplicates(subset=["sub_name"])
health_keep_df
health_keep_df = health_keep_df[health_keep_df["top_name"] == "Health"].reset_index()
del health_keep_df["index"]
```

```python
filtered_sub_top = pd.concat([gov_keep_df, health_keep_df, natsec_keep_df, finance_keep_df, eco_keep_df]).reset_index(drop=True)
filtered_sub_top.to_csv("filtered_sub_top.csv")
```
